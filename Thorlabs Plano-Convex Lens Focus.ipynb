# LA5315

```python


focal_LA5315 = 20 * mm
diameter = 12.7 * mm

x0 = np.linspace(-diameter / 2, diameter / 2, 1024 * 4)
z0 = np.linspace(-250 * um, focal_LA5315 + 10 * mm, 1024*4)

wavelength = 25  * um


u0 = Scalar_source_X(x=x0, wavelength=wavelength)
u0.plane_wave(A=1, theta=0 * degrees)

LA5315 = Scalar_mask_XZ(x=x0, z=z0, wavelength=wavelength)
LA5315.incident_field(u0)
LA5315.lens_convergent(r0=(0, 0),
                              aperture=diameter / 1.1,
                              radius=(8.7* mm, -8000000 * mm),
                              thickness=4.3 * mm,
                              refraction_index=1.51,
                              angle=0 * degrees,
                              mask=(5 * mm, 3 + 0.1j))
print(focal_LA5315 )
LA5315.draw_refraction_index()
edge_matrix = LA5315.borders

```

```python
LA5315.BPM(verbose=False)
LA5315.draw(logarithm=False,
        
        draw_borders=True,
        edge_matrix=edge_matrix)
```

```python
x_focus, z_LA5315_focus = LA5315.search_focus()
```
